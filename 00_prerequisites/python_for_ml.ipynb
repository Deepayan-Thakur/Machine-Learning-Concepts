# Variables & Data Types

```python
# Basic data types
x = 10              # int
y = 3.14            # float
name = "ML Engineer"  # string
is_ready = True     # boolean

type(x), type(y), type(name), type(is_ready)
```

# Control Flow

```python
score = 85

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
else:
    grade = "C"

grade
```

# Loops

```python
for i in range(5):
    print(f"Iteration {i}")
```

# Functions

```python
def accuracy(correct, total):
    return correct / total

accuracy(90, 100)
```

# Data Structures

```python
# List
features = ["age", "salary", "experience"]

# Tuple
shape = (100, 3)

# Set
unique_labels = {0, 1}

# Dictionary
model_params = {"lr": 0.01, "epochs": 100}

features, shape, unique_labels, model_params
```

# Comprehensions

```python
squared = [x**2 for x in range(10)]
even_squared = {x: x**2 for x in range(10) if x % 2 == 0}

squared, even_squared
```

# Lambda & Map

```python
nums = [1, 2, 3, 4]

list(map(lambda x: x**2, nums))
```

# Error Handling

```python
try:
    x = int("abc")
except ValueError:
    print("Conversion failed")
```
