Pandas is used for data manipulation and analysis.

# Import & Series

```python
import pandas as pd

s = pd.Series([10, 20, 30])
s
```

# DataFrame

```python
df = pd.DataFrame({
    "age": [25, 30, 35],
    "salary": [50000, 60000, 70000]
})

df
```

# Indexing

```python
df["age"]
df.iloc[0]
```

# Boolean Filtering

```python
df[df["salary"] > 55000]
```

# Handling Missing Values

```python
df_missing = df.copy()
df_missing.loc[1, "salary"] = None

df_missing.fillna(df_missing.mean())
```

# GroupBy

```python
df["department"] = ["HR", "IT", "IT"]

df.groupby("department").mean()
```
