# NumPy Basics

NumPy is the backbone of numerical computing in Machine Learning.

### Topics Covered
- Arrays
- Vectorized Operations
- Broadcasting
- Indexing & Slicing
- Linear Algebra


# Import & Arrays

```python
import numpy as np

arr = np.array([1, 2, 3, 4])
arr
```

# Array Shapes

```python
matrix = np.array([[1, 2], [3, 4], [5, 6]])

matrix.shape
```

# Zeros, Ones, Random

```python
np.zeros((3, 3)), np.ones((2, 2)), np.random.rand(3)
```

# Vectorized Ops

```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])

x + y, x * y
```

# Broadcasting

```python
data = np.array([[1, 2, 3], [4, 5, 6]])
data + 10
```

# Indexing & Slicing

```python
data[0]
data[:, 1]
```

# Linear Algebra

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

np.dot(A, B)
```
