# Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

# Line Plot

```python
x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot")
plt.show()
```

# Scatter Plot

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

# Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

# Seaborn Plot

```python
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

## Why Data Visualization Matters in ML

Visualization is not decoration  it's **diagnostics**.

We use plots to:
- Understand data distributions
- Detect outliers & leakage
- Validate assumptions (linearity, normality)
- Debug models (bias/variance)
- Communicate results to stakeholders


```python
# Imports (Best Practice)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

# Line Plot (Time Series / Training Curves)

## Line Plots

Used for:
- Training vs validation loss
- Time-series data
- Metric trends across epochs


```python
# Line Plot (Multiple Lines)
epochs = np.arange(1, 21)
train_loss = np.exp(-epochs / 5)
val_loss = train_loss + np.random.normal(0, 0.02, size=len(epochs))

plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

# Scatter Plot (Correlation Analysis)

## Scatter Plots

Used for:
- Feature vs target relationship
- Detecting non-linearity
- Identifying clusters


```python
# Scatter Plot
np.random.seed(42)

x = np.random.rand(100)
y = 3 * x + np.random.randn(100) * 0.2

plt.scatter(x, y)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Feature vs Target")
plt.show()
```

# Histogram (Distribution Check)

## Histograms

Used to:
- Check normality
- Detect skewness
- Decide transformations (log, box-cox)


```python
# Histogram
data = np.random.lognormal(mean=0, sigma=1, size=1000)

plt.hist(data, bins=40)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Feature Distribution")
plt.show()
```

# KDE Plot (Better Distribution View)

## KDE (Kernel Density Estimation)

Smoothed alternative to histograms.


```python
# Seaborn KDE Plot
sns.kdeplot(data, fill=True)
plt.title("KDE Plot")
plt.show()
```

# Box Plot (Outlier Detection)

## Box Plots

Used to:
- Detect outliers
- Compare distributions across categories


```python
# Box Plot
df = pd.DataFrame({
    "department": ["IT", "IT", "HR", "HR", "Sales", "Sales"],
    "salary": [70, 75, 50, 52, 60, 120]
})

sns.boxplot(x="department", y="salary", data=df)
plt.title("Salary Distribution by Department")
plt.show()
```

# Violin Plot (Box + KDE)

## Violin Plots

Box plot + distribution shape.
Great for deep analysis.


```python
# Violin Plot
sns.violinplot(x="department", y="salary", data=df)
plt.title("Violin Plot of Salaries")
plt.show()
```

# Bar Plot (Aggregations)

## Bar Plots

Used for:
- Aggregated statistics
- Categorical comparisons


```python
# Bar Plot
avg_salary = df.groupby("department")["salary"].mean().reset_index()

sns.barplot(x="department", y="salary", data=avg_salary)
plt.title("Average Salary by Department")
plt.show()
```

# Count Plot (Class Imbalance)

## Count Plots

Critical for checking **class imbalance** in ML datasets.


```python
# Count Plot
labels = pd.Series([0, 0, 0, 1, 1, 1, 1, 1])

sns.countplot(x=labels)
plt.title("Class Distribution")
plt.show()
```

# Heatmap (Correlation Matrix)

## Heatmaps

Used for:
- Correlation analysis
- Feature selection
- Multicollinearity detection


```python
# Correlation Heatmap
data = pd.DataFrame(np.random.rand(100, 5), columns=list("ABCDE"))

corr = data.corr()

sns.heatmap(corr, annot=True)
plt.title("Feature Correlation Matrix")
plt.show()
```

# Pair Plot (EDA Power Tool)

## Pair Plots

Visualizes:
- Feature distributions
- Pairwise relationships


```python
# Pair Plot
sns.pairplot(data)
plt.show()

```

# ML-Specific Visualization

## ML-Specific Plots

These plots are essential during model evaluation:
- Residual plots
- Prediction vs Actual


```python
# Prediction vs Actual Plot
y_true = np.random.rand(100)
y_pred = y_true + np.random.normal(0, 0.05, size=100)

plt.scatter(y_true, y_pred)
plt.plot([0, 1], [0, 1])
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Prediction vs Actual")
plt.show()
```
