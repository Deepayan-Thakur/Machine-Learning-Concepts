# Data Visualization with Matplotlib & Seaborn

Visualization helps understand data patterns and model behavior.

### Topics Covered
- Line, Bar, Scatter Plots
- Histograms
- Seaborn Statistical Plots


# Imports

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

sns.set()
```

# Line Plot

```python
x = np.arange(0, 10)
y = x ** 2

plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Line Plot")
plt.show()
```

# Scatter Plot

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

# Histogram

```python
data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram")
plt.show()
```

# Seaborn Plot

```python
tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

## Why Data Visualization Matters in ML

Visualization is not decoration  it's **diagnostics**.

We use plots to:
- Understand data distributions
- Detect outliers & leakage
- Validate assumptions (linearity, normality)
- Debug models (bias/variance)
- Communicate results to stakeholders


```python
# Imports (Best Practice)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)
```

# Line Plot (Time Series / Training Curves)

## Line Plots

Used for:
- Training vs validation loss
- Time-series data
- Metric trends across epochs


```python
# Line Plot (Multiple Lines)
epochs = np.arange(1, 21)
train_loss = np.exp(-epochs / 5)
val_loss = train_loss + np.random.normal(0, 0.02, size=len(epochs))

plt.plot(epochs, train_loss, label="Train Loss")
plt.plot(epochs, val_loss, label="Validation Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

# Scatter Plot (Correlation Analysis)

## Scatter Plots

Used for:
- Feature vs target relationship
- Detecting non-linearity
- Identifying clusters


```python
# Scatter Plot
np.random.seed(42)

x = np.random.rand(100)
y = 3 * x + np.random.randn(100) * 0.2

plt.scatter(x, y)
plt.xlabel("Feature")
plt.ylabel("Target")
plt.title("Feature vs Target")
plt.show()
```
